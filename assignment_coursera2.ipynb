Code to import libraries and reading the data from the webpage and converting it into a dataframe

```python
import pandas as pd
url  = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df1 = pd.read_html(url)
df1[0]
```

Code to remove the tuples in which the Borough value is 'Not assigned'

```python
df = df1[0][df1[0]['Borough'] != 'Not assigned']
df
```

To set the Neighbour value equal to Borough value in case Neighbour value in a tuple is 'Not assigned'

```python
df[df['Neighbourhood'] == 'Not assigned']
```

Now using the .loc we change 'Not assigned' to the value equal to the Borough value

```python
df.loc[7,'Neighbourhood'] = df.loc[7,'Borough']
df
    
```

Now we groupby the dataframe by the column "Postcode" and comma separate the values of neighbours for same value of 'Postcode'

```python
df = df.groupby(['Postcode','Borough'],as_index=False)['Neighbourhood'].agg(','.join)
df
```

Using the .shape method to print the number of rows of  dataframe

```python
df.shape
```

Converting the data in 'Geospatial_Coordinates.csv' into dataframe

```python
df2 = pd.read_csv('Geospatial_Coordinates.csv')
df2
```

Joining the two dataframes by .join method

```python
df3 = (df).join(df2)
df3
```

Droping the column 'Postal Code' additionally present to get the required dataframe

```python
df3.drop('Postal Code',axis = 1,inplace=True)
df3
```

Importing library for visulizing data in  map

```python
import folium
```

```python
toronto_map = folium.Map(location = [43.651070,-79.347015],zoom_start=10)
for lat,lng,label in zip(df3['Latitude'],df3['Longitude'],df3['Borough']):
    label = folium.Popup(label,parse_html = True)
    folium.CircleMarker(
    [lat,lng],
    radius = 1,
    popup = label,
    color = 'blue',
    fill = True,
    fill_color = '#3186cc',
    fill_opacity = 0.7,
    ).add_to(toronto_map)
toronto_map
```
